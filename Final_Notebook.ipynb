### This plot shows a no of churners and non churners we can observe count of churners around 500, and non churners more than 2500 

```python
#labels for pie chart
lab = telecom1["Churn"].value_counts().keys()
#values for pie chart
val = telecom1["Churn"].value_counts().values
list_of_tuples = list(zip(lab, val))# merging 2 list to single list 
per_plot=pd.DataFrame(list_of_tuples,columns=['a','b'])
per_plot.groupby(['a']).sum().plot(kind='pie', y='b', autopct='%1.0f%%',title='Percentage of churned and non-churned users ')
```

# Observation:

### From above calculations we can see that our almost 14% customer has been churned. We can definitely do something about that after analyzing the data.

```python
#Compare Churn with State using countplot()
plt.figure(figsize=(15,8))
sns.countplot(x=telecom1.State.sort_values(),hue='Churn',data=telecom1)
```

# Relation between Churn and Customer service

```python
#Calculating the Customer service calls with Churn percentage
Customer_service_percentage = pd.crosstab(telecom1['Customer service calls'],telecom1["Churn"])
Customer_service_percentage['Percentage_Churn'] = Customer_service_percentage.apply(lambda x : x[1]*100/(x[0]+x[1]),axis = 1)
Customer_service_percentage
```

```python
# Customer Calls
fig, axz = plt.subplots(figsize=(10,5))

axz = sns.countplot(x='Customer service calls', hue='Churn', data=telecom1, palette='Reds')

axz.set_title('Relation between churn and Customer service call', fontsize=20)
axz.set_ylabel('COUNTS', rotation=0, labelpad=50,size=10)
axz.set_xlabel('Customer Service Calls', size=10)
axz.yaxis.set_label_coords(-0.05, 0.95)  # (x, y)
axz.legend(loc=0,fontsize=20);

axz.tick_params(labelsize=15)  # Changes size of the values on the label
```

##Predictions
